```python
x= finaldata.drop(columns=["Alzheimer’s Diagnosis"] , axis=1) 
y = data["Alzheimer’s Diagnosis"] 
```

```python
from sklearn.model_selection import train_test_split 
xtrain,xtest , ytrain ,ytest = train_test_split(x,y , train_size=.8,random_state= 42) 
```

```python
from tpot import TPOTClassifier 
```

```python
model = TPOTClassifier(generations=5, population_size=50, cv=5, scoring='accuracy', verbosity=2, random_state=1, n_jobs=-1)
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.predict(xtest)
```
